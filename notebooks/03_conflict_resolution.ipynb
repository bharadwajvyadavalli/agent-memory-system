# ASMR: Conflict Resolution Deep Dive

This notebook explores the different types of conflicts ASMR can detect and resolve.

```python
import sys
sys.path.insert(0, '..')

from datetime import datetime, timedelta
from memory.schema import Memory, ConflictRecord
```

## Conflict Types

ASMR recognizes three types of conflicts:

1. **Full Contradiction**: Direct conflict where only one can be true
2. **Partial Update**: One memory updates part of another
3. **Scope Difference**: Different scopes, not actually conflicting

## 1. Full Contradiction

```python
now = datetime.utcnow()

# Full contradiction: CEO change
mem_a = Memory(
    id="ceo_old",
    content="John Smith is the CEO of TechCorp.",
    source="annual_report_2022",
    timestamp=now - timedelta(days=400),
)

mem_b = Memory(
    id="ceo_new",
    content="Jane Doe is the CEO of TechCorp (appointed March 2024).",
    source="press_release_2024",
    timestamp=now - timedelta(days=30),
)

# This would be detected as:
conflict = ConflictRecord(
    memory_a_id=mem_a.id,
    memory_b_id=mem_b.id,
    conflict_type="full_contradiction",
    winner_id=mem_b.id,
    reasoning="Both claim to identify the current CEO, but name different people. "
              "Memory B is newer (2024 vs 2022) and explicitly mentions appointment date.",
)

print("FULL CONTRADICTION")
print("="*50)
print(f"Memory A: {mem_a.content}")
print(f"Memory B: {mem_b.content}")
print(f"\nConflict type: {conflict.conflict_type}")
print(f"Winner: {conflict.winner_id}")
print(f"Reasoning: {conflict.reasoning}")
```

## 2. Partial Update

```python
# Partial update: Policy change affecting subset
mem_a = Memory(
    id="hours_general",
    content="Office hours are 9 AM to 5 PM, Monday through Friday.",
    source="employee_handbook",
    timestamp=now - timedelta(days=365),
)

mem_b = Memory(
    id="hours_remote",
    content="Remote employees can work flexible hours between 7 AM and 7 PM.",
    source="remote_policy",
    timestamp=now - timedelta(days=60),
)

conflict = ConflictRecord(
    memory_a_id=mem_a.id,
    memory_b_id=mem_b.id,
    conflict_type="partial_update",
    winner_id=None,  # Both valid for different contexts
    reasoning="Memory A describes general office hours, Memory B describes an exception "
              "for remote employees. Both are valid depending on the employee type.",
)

print("PARTIAL UPDATE")
print("="*50)
print(f"Memory A: {mem_a.content}")
print(f"Memory B: {mem_b.content}")
print(f"\nConflict type: {conflict.conflict_type}")
print(f"Winner: {conflict.winner_id} (both valid)")
print(f"Reasoning: {conflict.reasoning}")
```

## 3. Scope Difference

```python
# Scope difference: Not actually conflicting
mem_a = Memory(
    id="budget_company",
    content="Company annual budget is $10 million.",
    source="finance_report",
    timestamp=now - timedelta(days=30),
)

mem_b = Memory(
    id="budget_engineering",
    content="Engineering department budget is $3 million.",
    source="finance_report",
    timestamp=now - timedelta(days=30),
)

conflict = ConflictRecord(
    memory_a_id=mem_a.id,
    memory_b_id=mem_b.id,
    conflict_type="scope_difference",
    winner_id=None,  # Both valid, different scopes
    reasoning="No actual conflict. Memory A is company-wide, Memory B is department-specific. "
              "Engineering budget ($3M) is a subset of total ($10M).",
)

print("SCOPE DIFFERENCE")
print("="*50)
print(f"Memory A: {mem_a.content}")
print(f"Memory B: {mem_b.content}")
print(f"\nConflict type: {conflict.conflict_type}")
print(f"Winner: {conflict.winner_id} (not a real conflict)")
print(f"Reasoning: {conflict.reasoning}")
```
